```python
import os
from dotenv import load_dotenv
from huggingface_hub import login

load_dotenv()
# We use HF_TOKEN_READ from our .env file
hf_token = os.getenv('HF_TOKEN_READ')
if hf_token:
    print('Successfully loaded token from .env')
    login(token=hf_token)
else:
    print('HF_TOKEN_READ not found in .env')
```
