# Groq Setup (Ultra-fast LPU Inference)
Note: This is for **Groq.com**, which is different from xAI's Grok.

```python
from typing import TypedDict, List
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI
from langgraph.graph import StateGraph, START, END
from dotenv import load_dotenv # used to store secret stuff like API keys or configuration values
```

```python
from dotenv import load_dotenv
import os
from langchain_openai import ChatOpenAI

# Load environment variables
load_dotenv()

# Initialize Groq using ChatOpenAI (OpenAI compatible API)
# Note: llama3-70b-8192 is deprecated. Using llama-3.3-70b-versatile
llm = ChatOpenAI(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROQ_API_KEY"),
    base_url="https://api.groq.com/openai/v1"
)

# Example invocation
try:
    response = llm.invoke("Hello Groq! Tell me a very short joke about fast chips.")
    print(response.content)
except Exception as e:
    print(f"Error: {e}")
```

```python
class AgentState(TypedDict):
    messages: List[HumanMessage]
```

```python
def process(state:AgentState) -> AgentState:
		response = llm.invoke(state["messages"])
		print(f"\nAI: {response.content}")
		return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("process", process)
graph.add_edge(START, "process")
graph.add_edge("process", END) 
agent = graph.compile()
```

```python
user_input= input("Enter: ")
```

```python
while user_input != 'exit':
	agent.invoke({"messages": [HumanMessage(content=user_input)]})
	user_input = input("Enter: ")
```
