# Exploratory Data Analysis (EDA)

This notebook focuses on understanding the dataset structure, missing values, and distribution of features. 

**Note:** Actual data preprocessing (Imputation, Scaling, Encoding) for machine learning models is now handled via **Scikit-Learn Pipelines** in the respective modeling notebooks (e.g., `00_Pipeline_Setup.ipynb`, `02_Linear_Regression_Sklearn.ipynb`). This is to correctly prevent **data leakage**.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

### 1️⃣ Load Data

```python
df = pd.read_csv("data/dataset.csv")
print(f"Shape: {df.shape}")
df.head()
```

### 2️⃣ Data Structure & Missing Values

```python
print("=== DATA INFO ===")
print(df.info())

print("\n=== MISSING VALUES ===")
print(df.isnull().sum())
```

### 3️⃣ Statistical Summary

```python
df.describe()
```

### 4️⃣ Visualizing Distributions

```python
df.hist(bins=50, figsize=(20, 15))
plt.show()
```

### 5️⃣ Correlation Matrix

```python
plt.figure(figsize=(12, 8))
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("feature Correlations")
plt.show()
```

### ⚠️ Data Leakage Prevention

We observed missing values in `total_bedrooms` and that features are on different scales.

**Previously**, we imputed and scaled the entire dataset here. However, this causes **Data Leakage** because the test set information leaks into the training process (e.g., the mean used for scaling would be influenced by test data).

**Solution**: We will NOT modify and save a `dataset_processed.csv` here. Instead, we use `sklearn.pipeline.Pipeline` in the modeling notebooks to fit the imputer and scaler **only on the training data**.
