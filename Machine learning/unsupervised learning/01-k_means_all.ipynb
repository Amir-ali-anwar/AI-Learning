# **K-means Clustering**

* K-mean clustering is a type of unsupervised learning algorithm.
* It works on unlabeled data. 
* It make clusters from unlabeled data.

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
#ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# centers = [[2, 2], [-2, -2], [5, -3], [0, 0]]
# cluster_std = [0.5, 1.5, 0.3, 0.8] 

# X, y = make_blobs(
#     n_samples=600,
#     n_features=2,
#     centers=centers,
#     cluster_std=cluster_std,
#     random_state=42
# )

# X, y = make_blobs(
#     n_samples=500,
#     n_features=2,
#     centers=3,
#     cluster_std=[0.5, 1.2, 0.8],  # different spread
#     random_state=42
# )
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)

kmeans = KMeans(n_clusters=4, random_state=42)
# fit the data to the KMeans algorithm
kmeans.fit(X)
labels = kmeans.labels_
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title('K-Means Clustering')
plt.show()
```
